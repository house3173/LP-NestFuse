```python
import numpy as np
import matplotlib.pyplot as plt

# Dữ liệu từ bảng
data = {
    "LP-NestFuse_Old": [2.4483, 0.5460, 7.1849, 7.1773, 0.1702, 0.4958, 0.8651],
    "LP-NestFuse": [2.8512, 0.5427, 7.1866, 7.2536, 0.1829, 0.5183, 0.9214],
    "NestFuse": [2.4761, 0.3587, 7.1825, 6.9247, 0.1758, 0.4586, 0.9167],
    "NestFuse_New": [2.5743, 0.3718, 7.1828, 6.9122, 0.1685, 0.4650, 0.9060],
}

metrics = ["MI", "QG", "PSNR", "EN", "SD", "ALI", "VIF"]

labels = list(data.keys())
values = np.array(list(data.values())).T  # Chuyển danh sách thành ma trận, mỗi cột là một chỉ số

# Chuẩn hóa dữ liệu theo từng cột (chỉ số)
normalized_data = []
for col in values:
    # norm = (col - min(col)) / (max(col) - min(col))
    # norm = 0.5 + (col - np.min(col)) * 0.5 / (np.max(col) - np.min(col))
    norm = col / max(col)
    normalized_data.append(norm)

# Chuyển đổi lại về định dạng phù hợp
normalized_data = np.array(normalized_data).T  # Mỗi hàng là một phương pháp

# Số lượng chỉ số (trục)
num_vars = len(metrics)

# Tạo biểu đồ radar
fig, ax = plt.subplots(subplot_kw=dict(polar=True), figsize=(8, 8))

# Vẽ các trục
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
angles += angles[:1]  # Đóng tròn

# Thiết lập trục và nhãn
ax.set_theta_offset(np.pi / 2)
ax.set_theta_direction(-1)
ax.set_xticks(angles[:-1], metrics)

# Các mức dữ liệu
ax.set_ylim(0, 1)

# Vẽ từng bộ dữ liệu
for label, data_row in zip(labels, normalized_data):
    data_row = list(data_row) + [data_row[0]]  # Đóng vòng
    ax.plot(angles, data_row, label=label)
    ax.fill(angles, data_row, alpha=0.25)

# Tiêu đề và chú thích
# ax.set_title("Dataset: TNO", size=15, color="black", y=1.1)
ax.legend(loc='upper right', bbox_to_anchor=(1.3, 1.1))
plt.tight_layout()
plt.show()
```
