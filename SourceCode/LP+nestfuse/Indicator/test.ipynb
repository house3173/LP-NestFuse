```python
def mannos_skarison_filter(image):
    # Chuyn i nh sang min tn s
    f_transform = fft2(image)
    f_shift = fftshift(f_transform)
    
    # Ly kch thc nh
    rows, cols = image.shape
    cx, cy = cols // 2, rows // 2  # Tm nh
    
    # To b lc Mannos-Skarison
    x = np.linspace(-cx, cx, cols)
    y = np.linspace(-cy, cy, rows)
    X, Y = np.meshgrid(x, y)
    R = np.sqrt(X**2 + Y**2) + 1e-5  # Trnh chia cho 0
    filter_ms = (2.6 * (0.0192 + 0.114 * R) * np.exp(-((0.0192 + 0.114 * R) ** 1.1)))
    
    # p dng b lc trong min tn s
    filtered = f_shift * filter_ms
    
    # Chuyn v min khng gian
    f_ishift = ifftshift(filtered)
    image_filtered = np.abs(ifft2(f_ishift))
    
    return filtered, image_filtered 

image1 = mri_mri_ct.copy()
image2 = fused_mri_ct.copy()

diff_image = np.abs(image1 - image2)
filtered, image_filtered = mannos_skarison_filter(diff_image)

similarity_1 = np.mean((diff_image / 255.0) ** 2)
similarity_2 = np.mean((image_filtered / 255.0) ** 2)
print("Similarity 1: ", similarity_1)
print("Similarity 2: ", similarity_2)

plt.figure(figsize=(15, 15))
plt.subplot(1, 5, 1); plt.imshow(image1, cmap='gray'); plt.title("MRI")
plt.subplot(1, 5, 2); plt.imshow(image2, cmap='gray'); plt.title("Fused")
plt.subplot(1, 5, 3); plt.imshow(diff_image, cmap='gray'); plt.title("Difference")
plt.subplot(1, 5, 4); plt.imshow(np.abs(filtered), cmap='gray'); plt.title("Filter")
plt.subplot(1, 5, 5); plt.imshow(image_filtered, cmap='gray'); plt.title("Filtered Difference")
plt.show()
```
